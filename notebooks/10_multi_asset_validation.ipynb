# 10 Multi-Asset Validation

本课验证 `band_1pct` 是否只在 SPY 上有效，还是在多个资产上也有类似效果。

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_PATH = PROJECT_ROOT / "src"
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

from quant_trading.market_data import download_ohlcv
from quant_trading.strategy_improvement import (
    evaluate_multi_asset_band_filter,
    plot_multi_asset_band_filter,
    summarize_multi_asset_band_filter,
)
```

## 1. 下载多资产数据

```python
symbols = ["SPY", "QQQ", "DIA", "IWM", "EFA", "TLT"]
data_by_symbol = {
    symbol: download_ohlcv(symbol=symbol, start="2005-01-01", auto_adjust=True)
    for symbol in symbols
}
{symbol: df.shape for symbol, df in data_by_symbol.items()}
```

## 2. 运行 baseline vs band_1pct

```python
results = evaluate_multi_asset_band_filter(
    data_by_symbol,
    band_pct=0.01,
    short_window=10,
    long_window=200,
    transaction_cost_bps=3.0,
    slippage_bps=2.0,
    commission_bps=1.0,
)
results
```

## 3. 看相对 baseline 的改善

```python
comparison = summarize_multi_asset_band_filter(results)
comparison
```

## 4. 图表

```python
plot_multi_asset_band_filter(comparison);
```
