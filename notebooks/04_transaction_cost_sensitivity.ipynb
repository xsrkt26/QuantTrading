# 04 Transaction Cost Sensitivity

第四课目标：理解交易成本如何吃掉策略收益，以及为什么高换手策略更危险。

````python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC_PATH = PROJECT_ROOT / "src"
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

from quant_trading.market_data import download_ohlcv
from quant_trading.validation import (
    evaluate_transaction_cost_sensitivity,
    format_cost_sensitivity_table,
    plot_cost_sensitivity,
)
````

## 1. 成本假设

`bps` 是 basis points，基点。

```text
1 bps = 0.01%
10 bps = 0.10%
100 bps = 1.00%
```

本课测试同一个 `10/200` 双均线策略，在不同单边交易成本下的表现。

````python
SYMBOL = "SPY"
SHORT_WINDOW = 10
LONG_WINDOW = 200
COSTS_BPS = [0, 1, 5, 10, 25, 50]

df = download_ohlcv(symbol=SYMBOL, start="2000-01-01", auto_adjust=True)
results = evaluate_transaction_cost_sensitivity(
    df,
    short_window=SHORT_WINDOW,
    long_window=LONG_WINDOW,
    transaction_costs_bps=COSTS_BPS,
)

results
````

## 2. 看成本压力测试表

重点看：成本越高，最终净值、年化收益、Calmar 是否下降。

````python
print(format_cost_sensitivity_table(results))
````

## 3. 画图

如果成本稍微提高，策略就从盈利变亏损，这类策略执行风险很高。

````python
fig = plot_cost_sensitivity(results)
fig
````

## 4. 本课结论

`10/200` 双均线策略交易次数较少，所以成本敏感性不算极端。但成本从 0 bps 提高到 50 bps 后，最终净值仍明显下降。

未来研究任何策略，都要做成本压力测试。
