第一课目标：下载 SPY 日线行情，计算日收益率、净值曲线和回撤曲线。

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SRC_PATH = PROJECT_ROOT / "src"
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

from quant_trading.market_data import (
    add_return_equity_drawdown,
    download_ohlcv,
    format_summary,
    plot_price_equity_drawdown,
    summarize_performance,
)
```

## 1. 下载 SPY 日线数据

`auto_adjust=True` 表示使用复权价格，长期收益计算应优先使用复权价格。

```python
SYMBOL = "SPY"
START = "2000-01-01"

df = download_ohlcv(symbol=SYMBOL, start=START, auto_adjust=True)
df.head()
```

## 2. 从价格计算收益、净值和回撤

- `return`：今天收盘价 / 昨天收盘价 - 1
- `equity`：把每天收益率连续相乘，表示 1 块钱会变成多少
- `drawdown`：当前净值相对历史最高净值跌了多少

```python
analyzed = add_return_equity_drawdown(df)
analyzed[["Close", "return", "equity", "drawdown"]].tail()
```

## 3. 查看绩效摘要

```python
summary = summarize_performance(analyzed, symbol=SYMBOL)
print(format_summary(summary))
```

## 4. 画三条关键曲线

第一条是价格曲线，第二条是买入持有净值曲线，第三条是回撤曲线。

```python
fig = plot_price_equity_drawdown(analyzed, symbol=SYMBOL)
fig
```
